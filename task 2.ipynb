```python
import pandas as pd
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'[^a-z ]', '', text)
    return text

df['clean_text'] = df['Ticket Description'].apply(clean_text)

df[['Ticket Description', 'clean_text']].head()
```

```python
df['Ticket Type'].value_counts()
```

```python
df['Ticket Priority'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(df['clean_text'])

X.shape
```

```python
y_category = df['Ticket Type']
y_priority = df['Ticket Priority']
```

```python
from sklearn.model_selection import train_test_split

# For Ticket Type (Category)
X_train, X_test, y_train_cat, y_test_cat = train_test_split(
    X, y_category, test_size=0.2, random_state=42
)

# For Priority
X_train_p, X_test_p, y_train_pri, y_test_pri = train_test_split(
    X, y_priority, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

category_model = LogisticRegression(max_iter=1000)
category_model.fit(X_train, y_train_cat)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred_cat = category_model.predict(X_test)

print("Category Accuracy:", accuracy_score(y_test_cat, y_pred_cat))
print("\nCategory Classification Report:\n")
print(classification_report(y_test_cat, y_pred_cat))
```

```python
priority_model = LogisticRegression(max_iter=1000)
priority_model.fit(X_train_p, y_train_pri)
```

```python
y_pred_pri = priority_model.predict(X_test_p)

print("Priority Accuracy:", accuracy_score(y_test_pri, y_pred_pri))
print("\nPriority Classification Report:\n")
print(classification_report(y_test_pri, y_pred_pri))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test_cat, y_pred_cat)
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Category Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def predict_ticket(text):
    cleaned = clean_text(text)
    vector = tfidf.transform([cleaned])
    
    category = category_model.predict(vector)[0]
    priority = priority_model.predict(vector)[0]
    
    return category, priority


# Example test
sample_ticket = "My internet is not working and the connection keeps dropping."

predict_ticket(sample_ticket)
```
